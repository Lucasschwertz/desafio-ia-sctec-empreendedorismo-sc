# Analise de estabelecimentos por municipio em Santa Catarina

## Objetivo da analise

Este notebook organiza a leitura do arquivo Excel enviado pelo usuario, padroniza as colunas reais disponiveis e gera um ranking de municipios por quantidade de estabelecimentos. O foco e medir a representatividade de cada municipio no total e observar a concentracao dos maiores polos.


```python
from pathlib import Path

import pandas as pd

from src.clean import normalize_columns, basic_missing_treatment
from src.plots import bar_plot

RAW_PATH = "data/raw/dados_mapa_empresas.xlsx"
RANKING_PATH = "outputs/tabelas/ranking_municipios.csv"
PLOT_PATH = "outputs/graficos/top_15_municipios.png"
```

```python
df = pd.read_excel(RAW_PATH)
df.head()
```

```python
df = normalize_columns(df)
df = basic_missing_treatment(df)
df.columns = (
    df.columns.str.replace("[", "", regex=False)
    .str.replace("]", "", regex=False)
    .str.strip("_")
)
df = df.rename(columns={"munic\u00edpio": "municipio", "estabelecimentos": "estabelecimentos"})
df = df[["municipio", "estabelecimentos"]].copy()
df["municipio"] = df["municipio"].astype(str).str.strip()
df[["cidade", "uf"]] = df["municipio"].str.rsplit(" - ", n=1, expand=True)
df["cidade"] = df["cidade"].astype(str).str.strip()
df["uf"] = df["uf"].astype(str).str.strip()
df["estabelecimentos"] = pd.to_numeric(df["estabelecimentos"], errors="coerce")
df = df.dropna(subset=["estabelecimentos"])
df = df.sort_values("estabelecimentos", ascending=False).reset_index(drop=True)
df.head()
```

```python
total_estabelecimentos = df["estabelecimentos"].sum()
ranking = df.copy()
ranking["percentual_total"] = (ranking["estabelecimentos"] / total_estabelecimentos) * 100
top_20_municipios = ranking.head(20).copy()
participacao_acumulada_top_10 = ranking.head(10)["percentual_total"].sum()
top_20_municipios
```

## Interpretacao dos resultados

O ranking ordena os municipios pela quantidade de estabelecimentos em ordem decrescente. O percentual total mostra quanto cada municipio representa dentro do conjunto analisado, permitindo identificar o peso relativo dos principais centros urbanos.


```python
Path("outputs/tabelas").mkdir(parents=True, exist_ok=True)
Path("outputs/graficos").mkdir(parents=True, exist_ok=True)

ranking.to_csv(RANKING_PATH, index=False, encoding="utf-8")
bar_plot(
    ranking,
    x="cidade",
    y="estabelecimentos",
    title="Top 15 municipios por estabelecimentos",
    outpath=PLOT_PATH,
    top_n=15,
)

participacao_acumulada_top_10
```

## Observacao sobre concentracao economica

A participacao acumulada dos 10 primeiros municipios e um indicador direto de concentracao. Quanto maior esse valor, maior a dependencia do total em poucos polos municipais, o que sugere distribuicao economica mais concentrada no territorio analisado.
