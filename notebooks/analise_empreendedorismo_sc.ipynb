# Analise de Empreendedorismo em Santa Catarina

## Objetivo da analise

Este notebook analisa a distribuicao de estabelecimentos por municipio em Santa Catarina a partir de dados do Mapa de Empresas. O fluxo organiza a leitura do Excel, padroniza as colunas reais da base, separa cidade e UF e gera um ranking com metricas de participacao percentual e graficos de apoio.


```python
from pathlib import Path
import sys

import pandas as pd

PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.clean import normalize_columns, basic_missing_treatment
from src.plots import bar_plot

RAW_PATH = PROJECT_ROOT / "data" / "raw" / "dados_mapa_empresas.xlsx"
RANKING_PATH = PROJECT_ROOT / "outputs" / "tabelas" / "ranking_municipios.csv"
TOP_15_PATH = PROJECT_ROOT / "outputs" / "graficos" / "top_15_municipios.png"
TOP_10_PARTICIPATION_PATH = PROJECT_ROOT / "outputs" / "graficos" / "participacao_top10.png"
```

```python
df = pd.read_excel(RAW_PATH)
df.head()
```

```python
df = normalize_columns(df)
df = basic_missing_treatment(df)
df.columns = (
    df.columns.str.replace("[", "", regex=False)
    .str.replace("]", "", regex=False)
    .str.strip("_")
)
df = df.rename(columns={"munic\u00edpio": "municipio", "estabelecimentos": "estabelecimentos"})
df = df[["municipio", "estabelecimentos"]].copy()
df["municipio"] = df["municipio"].astype(str).str.strip()
df[["cidade", "uf"]] = df["municipio"].str.rsplit(" - ", n=1, expand=True)
df["cidade"] = df["cidade"].astype(str).str.strip()
df["uf"] = df["uf"].astype(str).str.strip()
df["estabelecimentos"] = pd.to_numeric(df["estabelecimentos"], errors="coerce")
df = df.dropna(subset=["estabelecimentos"])
df = df.sort_values("estabelecimentos", ascending=False).reset_index(drop=True)
df.head()
```

```python
total_estabelecimentos = df["estabelecimentos"].sum()
ranking = df.copy()
ranking["percentual_total"] = ((ranking["estabelecimentos"] / total_estabelecimentos) * 100).round(2)
top_20_municipios = ranking.head(20).copy()
top_10_participacao = ranking.head(10)[["cidade", "percentual_total"]].copy()
participacao_acumulada_top_10 = top_10_participacao["percentual_total"].sum().round(2)
top_20_municipios
```

## Interpretacao dos resultados

O ranking ordena os municipios pela quantidade de estabelecimentos em ordem decrescente. O percentual total mostra quanto cada municipio representa dentro do conjunto analisado, permitindo identificar os principais polos economicos do estado.


```python
RANKING_PATH.parent.mkdir(parents=True, exist_ok=True)
TOP_15_PATH.parent.mkdir(parents=True, exist_ok=True)

ranking.to_csv(RANKING_PATH, index=False, encoding="utf-8")
bar_plot(
    ranking,
    x="cidade",
    y="estabelecimentos",
    title="Top 15 municipios por estabelecimentos",
    outpath=str(TOP_15_PATH),
    top_n=15,
)
bar_plot(
    top_10_participacao,
    x="cidade",
    y="percentual_total",
    title="Participacao percentual dos 10 principais municipios",
    outpath=str(TOP_10_PARTICIPATION_PATH),
    top_n=10,
)
participacao_acumulada_top_10
```

## Observacao sobre concentracao economica

A participacao acumulada do top 10 ajuda a medir o nivel de concentracao economica. Quanto maior esse indicador, maior o peso de poucos municipios sobre o total de estabelecimentos observados na base.
